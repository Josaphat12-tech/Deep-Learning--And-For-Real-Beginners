### IMPORT LIBRARIES

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications import imagenet_utils
from tensorflow.keras.preprocessing.image import img_to_array,load_img
import numpy as np
import cv2
```

### LOAD THE IMAGE 

```python
img_path = '/content/download.jpg'
img = load_img(img_path)
disp_img = cv2.imread(img_path)
# cv2.imshow("actual image", disp_img)

%matplotlib inline
#The line above is necesary to show Matplotlib's plots inside a Jupyter Notebook

# import cv2
# from matplotlib import pyplot as plt

# #Import image
# image = cv2.imread('/content/download.jpg')

# #Show the image with matplotlib
# plt.imshow(image)
# plt.show()



# load image using cv2....and do processing.
plt.axis('off')
plt.imshow(cv2.cvtColor(disp_img, cv2.COLOR_BGR2RGB))
# as opencv loads in BGR format by default, we want to show it in RGB.
plt.show()
```

### RESIZE THE IMAGE AND CONVERT IT INTO AN ARRAY

```python

img = img.resize((224,224))
img_array = img_to_array(img)
#convert the image into 4D
img_array = np.expand_dims(img_array, axis=0)
```

### PREPROCESS THE INPUT IMAGE ARRAY 

```python
img_array = imagenet_utils.preprocess_input(img_array)
```

### LOAD THE MODEL

```python
pretrained_model = ResNet50(weights='imagenet')

#predict using predict()
prediction = pretrained_model.predict(img_array)

#decode the prediction
actual_prediction = imagenet_utils.decode_predictions(prediction)

print("predicted object is")
print(actual_prediction[0][1])
print(actual_prediction[0][4])
```
