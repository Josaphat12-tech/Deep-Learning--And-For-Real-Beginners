```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = r'https://raw.githubusercontent.com/codebasics/deep-learning-keras-tf-tutorial/master/6_gradient_descent/insurance_data.csv'
```

```python
df = pd.read_csv(url)                  #But if there was some comma i'd use pd.read_csv(csv_File,error_bad_lines=False)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df[['age','affordibility']],df.bought_insurance,test_size = 0.2,random_state = 25)
```

```python
len(x_train)
```

```python
df.shape
```

```python
# We are Scaling these Two Variables
x_train_scaled = x_train.copy()
x_train_scaled['age'] = x_train_scaled['age'] / 100        # For scaling we are just deviding these values to 100 since they are ages

x_test_scaled = x_test.copy()
x_test_scaled['age'] = x_test_scaled['age'] / 100          # For scaling we are just deviding these values to 100 since they are ages


# WE HAVE DONE THIS UP TO BRING BOTH VARIABLES AT THE  SCALE WHICH 0 AND 1
```

```python
x_train_scaled.head()
```

```python
model = keras.Sequential([
                  keras.layers.Dense(1,input_shape = (2,),activation='sigmoid',kernel_initializer='ones',bias_initializer='zeros')
])

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)


model.fit(x_train_scaled,y_train,epochs=5000)
```

```python
model.evaluate(x_test_scaled,y_test)
```
