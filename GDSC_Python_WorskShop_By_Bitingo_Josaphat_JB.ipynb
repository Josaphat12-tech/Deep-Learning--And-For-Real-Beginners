#**Project 1: Body Mass Index (BMI) Calculator**

```python
# BMI Calculator - Calculate the Body Mass Index (BMI) based on user input

# Prompt the user for their weight in kilograms
weight = float(input("Enter your weight in kilograms: "))

# Prompt the user for their height in meters
height = float(input("Enter your height in meters: "))

# Calculate the BMI using the formula: BMI = weight / (height * height)
bmi = weight / (height * height)

# Print the calculated BMI
print("Your BMI is:", bmi)

# Determine the BMI category based on the calculated value
if bmi < 18.5:
    print("Underweight")
elif 18.5 <= bmi < 25:
    print("Normal weight")
elif 25 <= bmi < 30:
    print("Overweight")
else:
    print("Obese")
```

#**Project 2: Password Strength Checker**

```python
# Password Strength Checker - Evaluate the strength of a password based on certain criteria

# Prompt the user to enter a password
password = input("Enter your password: ")

# Initialize a variable to keep track of the password strength score
strength = 0

# Check the length of the password and increment the strength score accordingly
if len(password) >= 8:
    strength += 1

# Check if the password contains uppercase letters and increment the strength score accordingly
if any(char.isupper() for char in password):
    strength += 1

# Check if the password contains lowercase letters and increment the strength score accordingly
if any(char.islower() for char in password):
    strength += 1

# Check if the password contains digits and increment the strength score accordingly
if any(char.isdigit() for char in password):
    strength += 1

# Check if the password contains special characters and increment the strength score accordingly
if any(not char.isalnum() for char in password):
    strength += 1

# Print the password strength based on the calculated score
if strength == 0:
    print("Weak password")
elif strength <= 2:
    print("Medium password")
else:
    print("Strong password")
```

#**Python Currency Converter Script based Project**

```python
# Currency Converter - Convert an amount from one currency to another

# Prompt the user to enter the amount to be converted
amount = float(input("Enter the amount to be converted: "))

# Prompt the user to enter the source currency
source_currency = input("Enter the source currency: ")

# Prompt the user to enter the target currency
target_currency = input("Enter the target currency: ")

# Define the exchange rates (example rates used)
exchange_rates = {
    'USD': {'EUR': 0.85, 'GBP': 0.75},
    'EUR': {'USD': 1.18, 'GBP': 0.88},
    'GBP': {'USD': 1.34, 'EUR': 1.14}
}

# Convert the amount to the target currency
if source_currency in exchange_rates and target_currency in exchange_rates[source_currency]:
    converted_amount = amount * exchange_rates[source_currency][target_currency]
    print("Converted amount:", converted_amount, target_currency)
else:
    print("Invalid source or target currency")
```

#**Number Guessing Game**

```python

import random

# Generate a random number between 1 and 10
secret_number = random.randint(1, 10)

print("Welcome to the Number Guessing Game!")
print("I'm thinking of a number between 1 and 10. Can you guess it?")

# Initialize the number of attempts
attempts = 0

while True:
    try:
        # Prompt the user to guess the number
        guess = int(input("Enter your guess: "))

        # Increment the number of attempts
        attempts += 1

        # Compare the guess with the secret number
        if guess == secret_number:
            print("Congratulations! You guessed it right.")
            print("Number of attempts:", attempts)
            break
        elif guess < secret_number:
            print("Too low! Try again.")
        else:
            print("Too high! Try again.")

    except ValueError:
        print("Invalid input. Please enter a valid number.")

print("Thank you for playing!")
```
